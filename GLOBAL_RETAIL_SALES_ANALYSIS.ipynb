**Univariate Analysis** â†’ distribution of single variables (countplot, histogram, boxplot)

```python
#1.Countplot
plt.figure(figsize=(5,4))
sns.countplot(x='Product Category', data=df)
plt.title('Number of Orders by Product Category')
plt.show()
```

## Interpretation of  above chart
**1.Explain the Chart-**
The chart visualizes the distribution of sales across different product categories and ranks them in descending order based on sales volume.

**2.What is it saying-**
The chart indicates which product categories are sold most frequently and which categories have lower sales volume .This analysis enables the identification of high-performing categories that contribute significantly to overall revenue.

**3.What features used-**
 The Product Category feature is used in a countplot to visualize the frequency of orders across different product categories.

**4. Business Insight -**
 The analysis reveals sales performance variations among product categories. High-performing categories may require sustained inventory support, while low-performing categories may need targeted business strategies such as promotional offers, pricing optimization, or inventory rationalization to improve sales efficiency and demand.


```python
# 2.Histogram
plt.figure(figsize=(8,6))
sns.histplot(df['Sales Price'], bins=10, kde=True)
plt.xlabel('Sales Price')
plt.ylabel('No Of orders')
plt.title('Distribution of Sales Price')
plt.show()
```

## Interpretation of your above chart

**1.Explain the Chart-**
The histogram displays the distribution of sales prices by grouping individual sales transactions into different price intervals. It shows how frequently sales occur within each price range across the dataset.

**2. What is it saying-**
The chart highlights the sales price ranges with the highest and lowest number of transactions. It indicates whether customer purchases are concentrated in lower, mid-range, or higher price segments.

**3. Features Used-**
The Sales Price feature is used as the numerical variable in the histogram. The x-axis represents sales price ranges, while the y-axis shows the frequency (number of transactions) within each range.

**4.  Business Insight -**
This analysis reveals customer purchasing patterns and price sensitivity. A higher concentration of sales in lower or mid-price ranges suggests that customers are more price-conscious. These insights help businesses optimize pricing strategies, plan discounts effectively, and identify opportunities to introduce premium products or adjust prices to maximize revenue.


```python
# 3. BoxPlot
plt.figure(figsize=(6,4))
sns.boxplot(x='Product Category', y='Total Sales',data=df,showfliers= False )
plt.xlabel('product Category')
plt.ylabel('Total Sales')
plt.title(' Spread of Total Sales by product category')
plt.show()
```

## Interpretation of your above chart

**1.Explain the Chart-**
The boxplot visualizes the distribution of shipping charges by summarizing how shipping costs vary across all orders.


**2. What is it saying-**
The chart shows the typical shipping charges paid by customers and highlights the spread of shipping costs, representing high and low shipping charges and indicating variability in delivery costs.

**3. Features Used-**
The Shipping Charges feature is used as the numerical variable in this univariate boxplot, which illustrates the overall distribution and variability of shipping costs across all orders.

**4.  Business Insight -**
This analysis shows how shipping charges vary across orders. Stable costs indicate consistent pricing, while higher charges suggest increased delivery expenses for certain orders. These insights help businesses manage shipping costs and improve customer satisfaction.

**Bivariate Analysis** â†’ relation between two variables (scatterplot, barplot, correlation heatmap)

```python
# 4.Scatterplot
plt.figure(figsize=(5,4))
sns.scatterplot(x='Sales Price', y='Total Sales', data=df)
plt.xlabel('Sales Price')
plt.ylabel('Total Sales')
plt.title('Scatterplot of Sales Price vs Total Sales')
plt.show()

```

## Interpretation of  above chart

**1. Explain the Chart â€“**
The scatterplot visualizes the relationship between sales price and total sales by plotting individual transactions. Each point represents a combination of sales price and the corresponding total sales value.

**2. What is it saying â€“**
The chart shows how sales price is related to total sales. It highlights that some price ranges lead to higher sales, while others show varied sales performance.

**3. What features are used â€“**
The Sales Price feature is used on the x-axis, and the Total Sales feature is used on the y-axis to analyze their relationship in  the scatterplot.

**4.  Business Insight  â€“**
This analysis helps identify how changes in product pricing impact total sales performance. It supports better pricing decisions and helps businesses focus on price ranges that generate higher revenue.

```python
# 5.Barplot
plt.figure(figsize=(5,4))
sns.barplot(x='Buyer Gender', y='Total Sales', data=df)
plt.xlabel('Buyer Gender')
plt.ylabel('Total Sales')
plt.title('Barplot of Total Sales by Buyer Gender')
plt.show()
```

## Interpretation of  above chart

**1. Explain the Chart â€“**
The barplot compares the average total sales across different buyer genders by displaying sales values for each gender category.

**2. What is it saying â€“**
The chart shows differences in total sales between genders, indicating that certain buyer groups contribute more to overall sales than others.

**3. What features are used â€“**
The Buyer Gender feature is used as the categorical variable on the x-axis, while the Total Sales feature is used as the numerical variable on the y-axis in the Barplot.

**4.  Business Insight  â€“**
This analysis helps to understand gender-based purchasing behavior. It supports targeted marketing strategies, personalized offers, and better customer segmentation to improve overall sales performance.

```python
#6.correlation heatmap
correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

## Interpretation of  above chart

**1.Explain the Chart â€“**
 The heatmap shows how features are related to each other using colors and numbers.

**2.What is it saying â€“** It shows which features increase together and which move in opposite directions.

**3.What features are used â€“** All the features are included to see their relationships in the Heatmap.

**4. Business Insight  â€“** This helps us understand patterns and connections, so we can focus on important relationships.

```python
# 8.Pivot Table
pivot_table = pd.pivot_table(
    df,
    values='Total Sales',
    index='Buyer Gender',
    columns='Product Category',
    aggfunc='mean'
)

pivot_table

```

## Interpretation of  above chart

**1. Explain the Table â€“**
The pivot table summarizes the average total sales for each product category, grouped by buyer gender.

**2. What is it saying â€“**
It shows how spending differs between male and female buyers across various product categories, making it easy to compare performance at a glance.

**3. What features are used â€“**
Buyer Gender is used for grouping rows, Product Category for columns, and Total Sales to calculate average values.

**4. Business Insight â€“**
This analysis helps businesses identify which product categories perform better for each gender. These insights support targeted marketing, product positioning, and inventory planning to improve overall sales.

```python
#9.Grouped Analysis
grouped_df = df.groupby(['Buyer Gender', 'Product Category'])['Total Sales'].mean()
grouped_df
```

## Interpretation of  above chart

**1. Explain the Analysis â€“**
This grouped analysis calculates the average total sales for each combination of buyer gender and product category.

**2. What is it saying â€“**
It shows how different product categories perform for male and female buyers, highlighting variations in purchasing behavior.

**3. What features are used â€“**
Buyer Gender and Product Category are used to create groups, and Total Sales is used to compute the average sales for each group in the grouped analysis.

**4. Business Insight â€“**
This analysis helps businesses understand gender-based preferences across product categories. These insights can be used for targeted promotions, better inventory planning, and improved sales strategies.

```python
# 10.Grouped Bar Chart
grouped_data = df.groupby(
    ['Product Category', 'Buyer Gender']
)['Total Sales'].mean().unstack()

grouped_data.plot(kind='bar', figsize=(8,5))
plt.xlabel('Product Category')
plt.ylabel('Average Total Sales')
plt.title('Average Total Sales by Product Category and Buyer Gender')
plt.xticks(rotation=45)
plt.show()
```

## Interpretation of  above chart

**1. Explain the Chart â€“**
The chart compares average total sales across different product categories, split by buyer gender.

**2. What is it saying â€“**
It shows how purchasing behavior differs between male and female buyers for each product category. Some categories perform better with a specific gender, while others show similar sales across both.

**3. What features are used â€“**
Product Category is used on the x-axis, Buyer Gender is used to group the bars, and Average Total Sales is used to measure performance.

**4. Business Insight â€“**
This analysis helps to identify gender-based preferences in product categories. These insights can support targeted marketing, customized promotions, and better product placement to maximize sales.

```python
# 11.LineChart
age_sales = df.groupby('Buyer Age')['Total Sales'].mean()

plt.figure(figsize=(6,4))
plt.plot(age_sales.index, age_sales.values, marker='o')
plt.xlabel('Buyer Age')
plt.ylabel('Average Total Sales')
plt.title('Trend of Average Total Sales by Buyer Age')
plt.show()
```

## Interpretation of  above chart

**1. Explain the Chart â€“**
The line chart shows how the average total sales change across different buyer age groups.

**2. What is it saying â€“**
The chart highlights age groups that contribute more to total sales and shows how spending behavior varies with age.

**3. What features are used â€“**
Buyer Age is used to group customers, and Average Total Sales is calculated to understand spending trends across ages in the Line chart.

**4.  Business Insight  â€“**
This analysis helps to identify high-spending age groups and design targeted marketing strategies, personalized offers, and age-specific product recommendations to increase sales.
